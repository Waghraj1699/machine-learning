```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/kaggle/input/facebook-ads/Facebook_ads_Data.csv')
```

```python
import random
T = 200
num_ads = 10
ad_selected = []

num_of_reward_1 = [0] * num_ads
num_of_reward_0 = [0] * num_ads

total_reward = 0

for n in range(0,T):
    ad = 0
    max_random = 0
    for i in range(0,num_ads):
        
        random_beta = random.betavariate(num_of_reward_1[i] + 1 ,num_of_reward_0[i] + 1)
        
        if(random_beta > max_random):
            max_random = random_beta
            ad = i
        ad_selected.append(ad)
        reward = dataset.values[n,ad]
        
        if reward == 1:
            num_of_reward_1[ad]=num_of_reward_1[ad] + 1
        else:
            num_of_reward_0[ad]=num_of_reward_0[ad] + 1
        
        total_reward = total_reward + reward
```

```python
plt.hist(ad_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
