# **Importing the libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing the datasets**

```python
data_set = pd.read_csv('/content/ Covid_data.csv')
x = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

# **Handling Missing Data**

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan ,strategy = 'mean')

imputer.fit(x[:,0:1])
x[:,0:1] = imputer.fit_transform(x[:,0:1])
```

```python
print(x)
```

```python
imputer.fit(x[:,4:5])
x[:,4:5] = imputer.fit_transform(x[:,4:5])
```

```python
print(x)
```

# **Encoding Categorical Data**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[1])],remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

# **Splitting Data into Test set and Train set**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,X_test,y_train,Y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
print(x_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(Y_test)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x_train[:,3:4] = sc.fit_transform(x_train[:,3:4])
X_test[:,3:4] = sc.fit_transform(X_test[:,3:4])

x_train[:,6:] = sc.fit_transform(x_train[:,6:])
X_test[:,6:] = sc.fit_transform(X_test[:,6:])
```

```python
print(x_train)
```
